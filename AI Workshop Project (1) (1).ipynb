```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
project_model = LogisticRegression()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(project_df[['platelets']], project_df.DEATH_EVENT, train_size=0.8)
X_test
project_model.fit(X_train, Y_train)
X_test
Y_predicted = project_model.predict(X_test)
Y_predicted
project_model.predict_proba(X_test)
platelets = project_model.score(X_test, Y_test)
```

```python
platelets
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(project_df[['serum_creatinine']], project_df.DEATH_EVENT, train_size=0.8)
X_test
project_model.fit(X_train, Y_train)
X_test
Y_predicted = project_model.predict(X_test)
Y_predicted
project_model.predict_proba(X_test)
serum_creatinine = project_model.score(X_test, Y_test)
```
